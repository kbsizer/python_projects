```python
def factorial(n):
    print("factorial has been called with n = " + str(n))
    if n == 1:
        return 1
    else:
        res = n * factorial(n-1)
        print("intermediate result for ", n, " * factorial(" ,n-1, "): ",res)
        return res	

print(factorial(5))
```
...and a blob of JavaScript.
```javascript
function factorial(n) {
    if (n === 0 || n === 1) {
        return 1;  // 0! = 1! = 1
    }
    return n * factorial(n - 1);
}

factorial(3); // returns 6
```

## Cell 2
In the cell below, we're going to write a little turtle graphics Python 2 program.  If all goes well, it should generate a spiral.

````python
import turtle
colors=['red','purple','blue','green','yellow','orange']
t=turtle.Pen()
turtle.bgcolor('black')
for x in range(360):
    t.pencolor(colors[x%6])
    t.width(x/100+1)
    t.forward(x)
    t.left(59)
````

````python
# Simple Matplotlib scatter plot example
# Note: The "magic" %matplotlib inline statement above makes the resulting plot render in a notebook cell
import numpy as np
import matplotlib.pyplot as plt

N = 100
x = np.random.rand(N)                        # generate N random numbers between 0.0 and 1.0
y = np.random.rand(N)
colors = np.random.rand(N)                   # TODO: need to better understand how these numbers map to colors
area = np.pi * (25*np.random.rand(N))**2     # pi r^2  0 < r < 25

plt.scatter( x, y, s=area, c=colors, alpha=0.5)
plt.show()
````
